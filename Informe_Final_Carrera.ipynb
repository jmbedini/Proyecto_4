#### Valores atípicos
A continuación identificamos y trabajamos sobre nuestros outliers para que no distorsionen nuestro análisis.

```python
dataset_ML = data_ML.copy()
```

```python
dataset_ML = dataset_ML.resample('D', on = 'fecha').sum()
dataset_ML.head()
```

```python
plt.figure(figsize = (15,6))
plt.plot(dataset_ML.index, dataset_ML.cantidad_pasos)
plt.xlabel('Tiempo')
plt.ylabel('Cantidad de autos, por mes')
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.boxplot(data = dataset_ML, x = "cantidad_pasos")
plt.xlim(-5000,150000)
plt.title("Outliers")
plt.xlabel("Flujo Vehicular")
```

Como podemos observar tanto en el primer gráfico como en el gráfico de caja, existen outliers en nuestro dataset que pueden distorsionar nuestro análisis. A simple vista, los valores por debajo de 20.000 pasos de autos podemos considerarlos como atípicos. 

Vamos a utilizar el **método del IQR** para trabajarlos, ya que se adapta mejor a nuestros datos.

```python
Q1 = dataset_ML['cantidad_pasos'].quantile(0.25)
Q3 = dataset_ML['cantidad_pasos'].quantile(0.75)
IQR = Q3 - Q1 #Rango intercuartílico

limite_inf = (Q1) - (1.5 * IQR)
limite_sup = (Q3) + (1.5 * IQR)

print(Q1, Q3, IQR)
print("Límite Inferior: " + str(limite_inf))
print("Límite Superior: " + str(limite_sup))
```

Ahora si con certeza podemos decir que aquellos valores por debajo de 20.098 pasos de vehiculos, los definimos con outliers. 

Como en una serie de tiempo **no podemos eliminar los registros** para que nuestra serie cuente con todos los datos equiespaciados, vamos a proceder a transformarlos en NAN y rellenarlos con la función **'interpolate'.**

```python
dataset_ML.cantidad_pasos[dataset_ML.cantidad_pasos < limite_inf] = np.nan
dataset_ML.cantidad_pasos[dataset_ML.cantidad_pasos > limite_sup] = np.nan
```

```python
print(dataset_ML.isna().sum())
```

```python
dataset_ML = dataset_ML.interpolate() #interpotlate lo que hace es rellenar los NA
print(dataset_ML.isna().sum())
```

```python
plt.figure(figsize = (10,5))
sns.boxplot(data = dataset_ML, x = "cantidad_pasos")
plt.xlim(-5000,150000)
plt.title("Outliers")
plt.xlabel("Flujo Vehicular")
```

```python
plt.figure(figsize = (15,6))
plt.plot(dataset_ML.index, dataset_ML.cantidad_pasos)
plt.ylim(0)
plt.xlabel('Tiempo')
plt.ylabel('Cantidad de autos, por mes')
plt.show()
```

### Modelo Benchmark

#### Modelo Benchmark  (XGBoost Regressor)

En **Series de Tiempo** para poder utilizar modelos de Regresión de aprendizaje **Supervisado**, debemos 'romper' la temporalidad de la misma.

Es por esto que vamos a utilizar la estretegia del **"ONE-STEP PREDICTION"**.

Para predecir el tráfico del **primer trimestre del 2020** utilizaremos los **tres primeros años** para entrenar, **1 año** para evaluar y **1 año** para Testeo.

```python
data_XGB = dataset_ML.copy()
data_XGB = data_XGB.reset_index()
print(data_XGB.shape)
data_XGB.head()
```

Resampleamos nuestro Dataset en semanas para trabajarlo.

```python
data_XGB_semanal = data_XGB.resample('W', on = 'fecha').mean()
print(data_XGB_semanal.shape)
data_XGB_semanal.head()
```

```python
senial_XGB = data_XGB_semanal.cantidad_pasos
print(senial_XGB.shape)
senial_XGB.head()
```

En primer lugar, debemos definir cuántos valores hacia atrás tomaremos.

```python
look_back = 52 # usamos 1 año para evaluar
```

Creamos los `X` e `y` con ceros para luego llenar con los datos. 

```python
N = senial_XGB.shape[0]
print(N)
```

```python
N = senial_XGB.shape[0]
X = np.zeros((N - look_back - 1,look_back))
y = np.zeros(N - look_back - 1)
print(X.shape, y.shape)
```

Llenamos los `X` e `y`

```python
for i in range(X.shape[0]):
    X[i,:] = senial_XGB.iloc[i:i+look_back]
    y[i] = senial_XGB.iloc[i+look_back]
```

```python
print(X)
```

```python
print(y)
```

Hacemos el **Train-Test Split**

```python
N_train = 158 - look_back # cantidad de instancias que vamos a usar para entrenar. 
N_test = 12 #cantidad de puntos que vamos a usar para evaluar
N_total = N_train + N_test

length_total = N_train + N_test + look_back

X_train = X[:N_train,:]
y_train = y[:N_train]
print(X_train.shape, y_train.shape)

X_test = X[N_train:N_train+N_test,:]
y_test = y[N_train:N_train+N_test]
print(X_test.shape, y_test.shape, length_total) 
```

Instanciamos el modelo

```python
from xgboost import XGBRegressor
modelo_xgb = XGBRegressor(objective = 'reg:squarederror', random_state = 10)
```

Entrenamos el modelo

```python
modelo_xgb.fit(X_train, y_train)
```

Predecimos

```python
y_train_pred_xgb = modelo_xgb.predict(X_train)
y_test_pred_xgb = modelo_xgb.predict(X_test)
```

Evaluamos utilizando la métrica **RMSE**.

RMSE es una buena medida de la precisión con que el modelo predice la respuesta, y es el criterio más importante para ajustar si el propósito principal del modelo es la predicción.

Esta métrica es muy suceptible a los outliers, pero al haberlos trabajado considero que es la correcta para evaluar este modelo, ya que es mas precisa que MAE.

```python
from sklearn.metrics import mean_squared_error
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred_xgb))
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred_xgb))
print(f'RMSE Train: {rmse_train}')
print(f'RMSE Test: {rmse_test}')
```

```python
plt.figure(figsize=(16,8))
plt.plot(senial_XGB.index[:length_total], senial_XGB[:length_total],lw = 2, label = 'Serie Original')
plt.plot(senial_XGB.index[:N_train + look_back], senial_XGB[:N_train + look_back],'--', label = 'Usado para entrenar')
plt.plot(senial_XGB.index[look_back: look_back + N_train], y_train_pred_xgb,'-.',label = 'Predicho en Train')
plt.plot(senial_XGB.index[N_train+look_back: length_total], y_test_pred_xgb,'-.', lw = 4, label = 'Predicho en Test')
plt.ylim(50000)
plt.legend()
plt.xlabel('Tiempo')
plt.ylabel('Flujo Vehicular')
plt.show()
```

#### Comentarios:
- Podemos decir que la performance de nuestro modelo no es mala, sobre todo si observamos el resultado de nuestro set de entrenamiento (RMSE Train XGB: 612.76)
- Además observando el gráfico nuestra linea roja discontinua, podemos ver como se ajusta aunque no del todo exacto.
- A pesar de esto, vemos que hay un gran **sobreajuste** en el mismo, siendo el RMSE est: 10699.70).
  Este sobreajuste se puede resolver simplificando el modelo o agregando mas datos.
- Vamos a intentar mejorar los resultados del modelo mediante la **optimización de Hiperparámetros.**

#### Optimización de Hiperparámetros
Partimos de los mismos X_train y X_test del modelo anterior:

```python
print(X_train)
```

```python
print(y_train)
```

Instanciamos el modelo

```python
modelo_xgb = XGBRegressor(objective = 'reg:squarederror', random_state = 10)
```

La idea del **XGBoost**, es ir entrenando modelos simples y que cada uno vaya aprendiendo de los errores del anterior, por lo que, para seguir con la simplicidad de los modelos, no vamos a entrenar un XGBoost con profundidad mayor a 4 para y evitar la complejización del modelo.

```python
from sklearn.model_selection import GridSearchCV

parametros = {
             "max_depth" : [1, 2, 3, 4],
             "n_estimators" : [100, 150, 200],
             } 

gs_xgb_gscv = GridSearchCV(estimator = modelo_xgb, param_grid = parametros,
                           scoring = 'neg_root_mean_squared_error')
```

```python
gs_xgb_gscv.fit(X_train, y_train)
```

```python
print(gs_xgb_gscv.best_params_)
print("Mejor score: " + str(abs(gs_xgb_gscv.best_score_)))
```

```python
df_xgb_gscv = XGBRegressor(objective = 'reg:squarederror',
                           random_state = 10, max_depth = gs_xgb_gscv.best_params_['max_depth'], 
                           n_estimators = gs_xgb_gscv.best_params_['n_estimators'])
```

```python
df_xgb_gscv.fit(X_train, y_train)
```

```python
y_train_pred_xgb_gscv = df_xgb_gscv.predict(X_train)
y_test_pred_xgb_gscv = df_xgb_gscv.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred_xgb_gscv))
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred_xgb_gscv))
print(f'RMSE Train: {rmse_train}')
print(f'RMSE Test: {rmse_test}')
```

```python
plt.figure(figsize=(16,8))
plt.plot(senial_XGB.index[:length_total], senial_XGB[:length_total],lw = 2, label = 'Serie Original')
plt.plot(senial_XGB.index[:N_train + look_back], senial_XGB[:N_train + look_back],'--', label = 'Usado para entrenar')
plt.plot(senial_XGB.index[look_back: look_back + N_train], y_train_pred_xgb_gscv,'-.',label = 'Predicho en Train')
plt.plot(senial_XGB.index[N_train+look_back: length_total], y_test_pred_xgb_gscv,'-.', lw = 4, label = 'Predicho en Test')
plt.ylim(50000)
plt.legend()
plt.xlabel('Tiempo')
plt.ylabel('Flujo Vehicular')
plt.show()
```

### S.A.R.I.M.A.X


Antes que nada, me parece adecuado recordar para nuestras predicciones que estamos partiendo de un supuesto muy importante: **que el futuro se va a comportar como el pasado**.

Debemos trabajar con **SARIMAX** y no **ARIMA**, ya que es una Serie de Tiempo *no estacionaria y con estacionalidad*. 

   
   - **Procesos Autorregresivos AR(p)**: Se podría decir que es el grado en que el valor de hoy depende del mismo valor, pero del pasado. Sería cómo una regresión lineal de los **p** valores anteriores a al momento de hoy.
   
     
   - **Diferenciación(d)**: para poder trabajar con series de tiempo y modelos regresivos, necesitamos que sea **estacionaria**, por lo que creamos una diferenciación para lograrlo (podría decirse que es una normalización).
     
     
   - **Procesos de media móvil MA(q)**: se conforman como una combinación de variables aleatorias no correlacionadas, donde buscamos incluir en nuestro valor en un determinado momento,los distintos valores de **ruido blanco** que existieron a lo largo del tiempo. Sería como una regresión lineal de los **q** términos de error anteriores al momento de hoy.
   
   

- Cuando trabajamos con series de tiempor que tienen "efectos estacionales", el modelo seasonal A.R.I.M.A es el mas adecuado para utilizar, el cual se representa como `ARIMA(p,d,q)(P,D,Q)`.


* Donde `(p,d,q)` son los paramatros **no-estacionales** mencionados, mientras que `(P,D,Q)` siguen la misma definicion pero son aplicados a los parametros estacionales de la serie.

Importamos algunas librerias que vamos a necesitar:

```python
import statsmodels.tsa as tsa
import statsmodels.api as sm
from statsmodels.tsa import stattools
import warnings, itertools
import datetime as dt
import sklearn.metrics as metrics
```

Definimos los parámetros referidos a Estacionalidad, Tendencia y Ruido.

```python
p = q = range(6, 10)
d = range(0,2)

# Se generan todas las diferentes combinaciones
pdq = list(itertools.product(p, d, q))

# Se generan todas las diferentes combinaciones estacionarias
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
print("Ejemplos de combinaciones para -seasonal A.R.I.M.A")
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```python
data_SARIMAX = dataset_ML.copy()
data_SARIMAX = data_SARIMAX.reset_index()
print(data_SARIMAX.shape)
data_SARIMAX.head()
```

```python
data_SARIMAX_semanal = data_SARIMAX.resample('W', on = 'fecha').mean()
print(data_SARIMAX_semanal.shape)
data_SARIMAX_semanal.head()
```

Según Peterson, T. (2014) el **AIC (Akaike information criterion)** es un estimador de la calidad de los modelos según una dataset trabajdo. Dado diferentes modelos, el AIC estima la calidad de cada uno de esos modelos, comparandolos con los otros.

Fuente: https://www.sciencedirect.com/topics/medicine-and-dentistry/akaike-information-criterion#:~:text=Akaike%20information%20criterion%20(AIC)%20(,among%20all%20the%20other%20models.)

Por lo tanto un bajo valor de AIC se considera mejor.


Según nuestro índice de **AIC**, el resultado óptimo es de: SARIMAX(4, 0, 4)x(4, 2, 11, 12) - AIC:48.0


Dado que probando esta combinación, el resultado no es el mejor, decidí realizar una busqueda manual de combinaciones de los parámetros.

#### *Entrenando*

```python
y= data_SARIMAX_semanal.cantidad_pasos


model= sm.tsa.statespace.SARIMAX(y,
                                order=(10,1,10),
                                seasonal_order=(10, 1, 10, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)

results= model.fit()

print(results.summary().tables[1])
```

* _*Esta tabla devuelve los resultados de SARIMAX con informacion importante. Por ejemplo la columna **coef** representa los **pesos o importancia** de cada feature.*_

* _*Podemos ver que varios de ellos aportan significativamente a la performance del modelo.*_

### Predicción 1er Trimestre 2020

#### *Modelando One-Step*

* _*Comparo valores reales contra valores predichos de la serie de tiempo. La funcion `get_prediction` y `conf_int` permite obtener los valores e intervalos de confianza para la predicion.*_

```python
data_SARIMAX_semanal[(data_SARIMAX_semanal.index > '2020-01-01')&(data_SARIMAX_semanal.index < '2020-04-01')]
```

```python
prediction = results.get_prediction(start= dt.datetime(2020,1,5), dynamic=False)
pred_conf = prediction.conf_int()
```

* _*Al pasar como argumento False al parametro 'dynamic' me aseguro de estar haciendo una One-Step-Prediction. Esto significa que la prediccion echa para cada valor esta generada con "toda la historia disponible"(todos los datos) hasta ese punto.*_

```python
plt.figure(figsize=(16,8))

ax = y['2017':].plot(label='Serie Observada')
prediction.predicted_mean.plot(ax=ax, label='SARIMAX One-Step-Prediction', alpha=0.7)

ax.fill_between(pred_conf.index,
                pred_conf.iloc[:, 0],
                pred_conf.iloc[:, 1], color='k', alpha=0.2)

plt.title("SARIMAX One-Step 1er Trimestre 2020", fontsize=20)
plt.xlim(dt.datetime(2019,1,1), dt.datetime(2020,4,1))
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)

plt.legend(fontsize=15)
plt.show()
```

#### *Evaluando*

```python
y_forecast= prediction.predicted_mean
y_real = y['2020-1-05':]


rmse_sarimax_one = np.sqrt(metrics.mean_squared_error(y_forecast, y_real))
print("RMSE- SARIMAX: ", rmse_sarimax_one)
```
